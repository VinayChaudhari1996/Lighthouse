```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
chess = cv.imread("CHESS.webp")
image_copy = cv.cvtColor(chess,cv.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray = cv.cvtColor(image_copy,cv.COLOR_RGB2GRAY)

gray = np.float32(gray)

#Harris corner Detector required float32 data type
#Harris corner detector required gray scale float input image
dst = cv.cornerHarris(gray,2,3,0.04)

dst = cv.dilate(dst,None)

plt.imshow(dst,cmap='gray')
```

```python
thresh = 0.001*dst.max()
corner_image = np.copy(image_copy)
for i in range(0,dst.shape[0]):
    for j in range(0,dst.shape[1]):
        if dst[i,j]>thresh:
            cv.circle(corner_image,(i,j),2,(0,255,0),1)
plt.imshow(corner_image)
```
