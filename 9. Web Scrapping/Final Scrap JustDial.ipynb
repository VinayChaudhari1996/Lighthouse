```python
import cv2 as cv
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import re
import time
```

```python
selenium = webdriver.Chrome("chromodriver/chromedriver.exe")
selenium.get("https://www.justdial.com/Pune/Restaurant")   
soup = BeautifulSoup(selenium.page_source,"html.parser")
tags  = soup.find("div",attrs={'class':"jpag"})
page_links = ["https://www.justdial.com/Pune/Restaurants"]
for tag in tags.findAll("a")[1:-1]:
    page_links.append(tag['href'])
selenium.quit()
```

```python
page_links
```

```python
just_dial = {"name":[],"vote":[],"rate":[],'image':[],"address":[],'expensive':[],
             'dish_info':[],"amentes":[],"payment_mode":[],"phone_number":[]}

for page_link in page_links:
    selenium = webdriver.Chrome("chromodriver/chromedriver.exe")
    selenium.get(page_link)
    lenOfPage = selenium.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    match=False
    while(match==False):
        lastCount = lenOfPage
        time.sleep(3)
        lenOfPage = selenium.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
        if lastCount==lenOfPage:
            match=True
    source_data = selenium.page_source
    
    soup = BeautifulSoup(source_data,"html.parser")
    
    ul = soup.find("ul",attrs={"class":"rsl col-md-12 padding0"})
    lis = ul.find_all("li",attrs={"class":"cntanr"})
    print(lis.__len__(),lenOfPage)

    cipher_keys = str(soup.select("style[type='text/css']")[1])
    keys = re.findall("-(\w+):before",cipher_keys,flags=0)

    values = [int(item)-1 for item in re.findall("9d0(\d+)",cipher_keys,flags=0)]
    ciphers_dict = dict(zip(keys,values))
    j = 1
    for li in lis:
   
        image = li.find("img",attrs={'class':"altImgcls"})['src']
        rest_name = li.find("span",attrs={"class":"lng_cont_name"}).text
        rest_votes = li.find("span",attrs={'class':"rt_count lng_vote"}).text.strip().split("\t")[0]
        rest_rate = li.find("span",attrs={'class':"green-box"}).text

        main_href = li["data-href"]

        sel = webdriver.Chrome("chromodriver/chromedriver.exe")
        sel.get(main_href)
        beautful = BeautifulSoup(sel.page_source,'html.parser')
        try:
            rest_address = beautful.find("span",attrs={"class":"comp-text"}).text.strip()
        except Exception as error :
            rest_address = np.nan
        try:
            rest_expensive = beautful.find("span",attrs={"class":"rshover"}).text.strip()
        except Exception as error:
            rest_expensive = np.nan
        try:
            rest_dish_info = beautful.find("span",attrs={"class":"quickinfosub wrapper"}).text
        except Exception as error:
            rest_dish_info = np.nan
            print(error)
        try:
            rest_amentes = beautful.findAll("div",attrs={"class":"quickinfowrp wrapper"})[1].text
        except Exception as error:
            rest_amentes=np.nan
            print(error)
        try:
            payment_mode = beautful.findAll("div",attrs={'class':"mreinfwpr"})[2].text 
        except Exception as error:
            payment_mode=np.nan
        cipher_keys = str(beautful.select("style[type='text/css']")[1])
        keys = re.findall("-(\w+):before",cipher_keys,flags=0)

        values = [int(item)-1 for item in re.findall("9d0(\d+)",cipher_keys,flags=0)]
        ciphers_dict = dict(zip(keys,values))

        b = beautful.find("a",attrs={'class':"tel ttel"})
        
        b = str(b)
        c = re.findall(r'[\wmobilesv icon-]+[">]',b)
        q = []
        for i in c:
            if i.find("mobilesv")==0:
                q.append(i)
        
        number = []
        plus = "+"
        number.append(plus)
        for class_ in q[1:]:    
            key = class_.split("-")[-1]
            number.append(str(ciphers_dict[key[:-1]]))
        phone_number = "".join(number)

        just_dial["name"].append(rest_name)
        just_dial["vote"].append(rest_votes)
        just_dial["rate"].append(rest_rate)
        just_dial["image"].append(image)
        just_dial['address'].append(rest_address)
        just_dial["expensive"].append(rest_expensive)
        just_dial["dish_info"].append(rest_dish_info)
        just_dial["amentes"].append(rest_amentes)
        just_dial["payment_mode"].append(payment_mode)
        just_dial["phone_number"].append(phone_number)
        sel.quit()
    sel.quit()
    
```

```python
df = pd.DataFrame(just_dial)
df.to_csv("restaurant.csv")
```
